  Bên cạnh những người ý thức và chấp hành tốt các quy định phòng tránh dịch thì còn một số bộ phận còn lơ là, không chấp hành quy định. Còn tụ tập nhiều người, không giữ khoảng cách. Những trường như vậy cần phải được phát hiện và xử phạt.
  
  Khi bắt gặp một vài người vô tình, hay cố ý đứng gần nhau, thì cần phải xác định được xem những người đó đã tuân thủ quy định giữ đúng khoảng cách hay chưa? Lúc đó chúng ta cần chụp lại hình ảnh, phân tích, nếu không tuân thủ quy định thì cần báo cho cơ quan chức năng và y tế để giải quyết.

- Input: Một bức hình chụp một nhóm người
